```python
import math, numpy
#Part 1 input data: Air temp (degC), ID (W/sqm), load resistance (ohms)
# - split into training set and a randomly slected small validation set
xdata = [[-10.0, 350, 4.464], 
  [-10.0, 650, 4.464], 
  [-10.0, 950, 4.464], 
  [-10.0, 1250, 4.464], 
  [10.0, 350, 4.464], 
  [10.0, 650, 4.464], 
  [10.0, 950, 4.464], 
  [10.0, 1250, 4.464], 
  [30.0, 350, 4.464], 
  [30.0, 650, 4.464], 
  [30.0, 950, 4.464], 
  [30.0, 1250, 4.464], 
  [-10.0, 350, 6.696], 
  [-10.0, 650, 6.696], 
  [-10.0, 950, 6.696], 
  [-10.0, 1250, 6.696], 
  [10.0, 350, 6.696], 
  [10.0, 650, 6.696], 
  [10.0, 950, 6.696], 
  [10.0, 1250, 6.696], 
  [30.0, 350, 6.696], 
  [30.0, 650, 6.696], 
  [30.0, 950, 6.696], 
  [30.0, 1250, 6.696], 
  [-10.0, 350, 8.928], 
  [-10.0, 650, 8.928], 
  [-10.0, 950, 8.928], 
  [-10.0, 1250, 8.928], 
  [10.0, 350, 8.928], 
  [10.0, 650, 8.928], 
  [10.0, 950, 8.928], 
  [10.0, 1250, 8.928], 
  [30.0, 350, 8.928], 
  [30.0, 650, 8.928], 
  [30.0, 950, 8.928], 
  [30.0, 1250, 8.928]] 

xarray= numpy.array(xdata)
```

```python
MeanTair= np.mean(xarray[:,0])
MeanID= np.mean(xarray[:,1])
MeanRL= np.mean(xarray[:,2])
Meantotal=[MeanTair,MeanID,MeanRL]
Xstd=np.std(xarray)
X = (xarray - Meantotal)/Xstd
print(X)
```

```python
print (X.T)  #print the transpose
Xmean = np.mean(X,0)  # compute mean vector
print (Xmean)
```

```python
C = np.cov(X.T)  #transpose is matrix we want to work with - compute covariance matrix
print (C)
```

```python
w, v = LA.eig(C)  # get the eigenvalues w and the eigenvectors 
print (w)
print (v)
```

```python
A  = np.array([v[:,2],v[:,1]])  # form the transformation matrix using eigenvectors
# for the top two eigenvalues
print(A)
```

```python
Y = np.matmul(A,(X-Xmean).T) # Apply transformation to obtain new   
                             # data representation in 2-D
print(Y)
```

```python
xhat = np.matmul(A.T,Y).T + Xmean  #recover the original data using the inverse transform
print(xhat)
```

```python
mse = np.sum((X - xhat)**2)/10  #mean squared error
print(mse)
```
